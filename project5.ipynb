```python
import pandas as pd
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

def draw_plot():
    fig1, ax1 = plt.subplots(figsize=(10, 6))
    ax1.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], alpha=0.6)

    ax1.set_xlabel("Year")
    ax1.set_ylabel("Sea Level")
    ax1.set_title("Rise in Sea Level")

    return fig1

draw_plot()
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

def draw_plot():
    fig2, ax2 = plt.subplots(figsize=(10, 6))
    ax2.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], alpha=0.5)

    ax2.set_title("Rise in Sea Level")
    ax2.set_xlabel("Year")
    ax2.set_ylabel("Sea Level")

    slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

    years_extended = np.arange(1880, 2051)
    sea_levels = intercept + slope * years_extended
    ax2.plot(years_extended, sea_levels, 'r')

    ax2.legend()
    return fig2

draw_plot()
plt.show()
```

```python
def draw_plot():
    fig3, ax3 = plt.subplots(figsize=(10, 6))
    ax3.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], alpha=0.6)

    ax3.set_title("Rise in Sea Level")
    ax3.set_xlabel("Year")
    ax3.set_ylabel("Sea Level")

    slope, intercept, _, _, _ = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    years_extended = np.arange(1880, 2051)
    ax3.plot(years_extended, intercept + slope * years_extended, 'r')

    df_recent = df[df['Year'] >= 2000]
    slope_recent, intercept_recent, _, _, _ = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])

    years_recent = np.arange(2000, 2051)
    ax3.plot(years_recent, intercept_recent + slope_recent * years_recent, 'green', label="Best fit")

    ax3.legend()
    return fig3

draw_plot()
plt.show()
```
